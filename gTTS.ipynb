```python
from gtts import gTTS
from IPython.display import Audio

# Convert text to speech
tts = gTTS("My name is TASAWAR HUSSAIN and i am 51 year old")
tts.save("My name is TASAWAR HUSSAIN and i am 51 year old.mp3")

# Play the audio in Colab
Audio("My name is TASAWAR HUSSAIN and i am 51 year old.mp3")
```
